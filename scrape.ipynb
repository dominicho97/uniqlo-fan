```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.uniqlo.com/eu/en/men/featured/limited-offers"
```

```python
#Headers for request
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5)'
                    'AppleWebKit/537.36 (KHTML, like Gecko)'
                    'Chrome/50.0.2661.102 Safari/537.36'
}
```

```python
#HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
type(webpage.content)
```

```python
#Soup Object containing all data
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python
soup
```

```python
#Find all items from a -links
links = soup.find_all('a', attrs={'class':'productTile__link js-productLink'})
```

```python
links
```

```python
#get all item details href link
link = links[0].get('href')
```

```python
product_list = "https://www.uniqlo.com/" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
#Soup Object containing all data
new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
```

```python
new_soup
```

```python
new_soup.find("h1",attrs={"itemprop":"name"}).text
```
